## New York Property tax records - anomaly detection

Property tax records have entries that don't declare the proper dollar valuation commensurate to the area and size. Such anomalous records could be potential fraud declarations on the part of the owner to evade tax, this projects aims to detect such records.

### Goal

This project builds an unsupervised learning model to flag anomalous tax records for further review. An anomaly score is devised using the average of two measures:
    
1.) root sum of squares of the Euclidean distances from origin of the top dimensions of a PCA transformed feature set <br>
2.) autoencoder reproduction error of the top dimensions of a PCA transformed feature set

### Methodology

Impute records for missing values - cannot throw out records with missings as they may be the ones with anomalies. <br>
The trust implicit here is government owned buildings list the records with subterfuge, and thus excluded from study. <br>

2 imputation rules:
    
1.) records sorted by zip codes, ones with missing zip take the same zip if it matches the one before and after, or just the one before <br>
2.) all other missing fields filled in with mean values corresponding to its tax class

### Feature creation <br>

9 property valuations are created, by normalizing with size: <br>

Full Value, appraised value, land value each divided by building size (area), lot size (area) and building volume (area x #stories) <br>

For each zip, zip3 (1st 3 letters of zip), tax class and borough, 4 mean values of the 9 property valuations are estimated and joined on the primary key (zip, zip3, tax class, borough) to the records table. <br>

36 ratios are created with the 9 property valuations divided by each of the 4 mean value estimates: <br>

eg. for every row: (Full Value / lot size ) of a property at zip XXXXX    /   Mean (Full Value / lot size) of all properties at zip XXXXX <br>

9 property valuations and 36 ratios = 45 new features

```python
keep_list = ['RECORD', 'B', 'OWNER', 'TAXCLASS' , 'LTFRONT' , 'LTDEPTH', 'STORIES', 'FULLVAL', 'AVLAND', 'AVTOT', 'ZIP', 'BLDFRONT', 'BLDDEPTH']

property_data = df[keep_list].copy()

property_data['imputation'] = None # fill in details if row was imputed
```

```python
#Missing ZIPS
missing_zips = np.where(pd.isnull(property_data['ZIP']))[0]
print("Num records with missing zips: ", len(missing_zips))
```

```python
for i in missing_zips:
    if property_data.loc[i+1, 'ZIP'] == property_data.loc[i-1, 'ZIP']:
        property_data.loc[i, 'ZIP'] = property_data.loc[i-1, 'ZIP']
        property_data.loc[i, 'imputation'] = 'zip'
        
missing_zips = property_data[pd.isnull(property_data['ZIP'])].index
print("Number of missing zips after first round imputation:  ", len(missing_zips))
```

```python
for i in missing_zips:
    property_data.loc[i, 'ZIP'] = property_data.loc[i-1, 'ZIP']
    property_data.loc[i, 'imputation'] = 'zip'
    
missing_zips = np.where(pd.isnull(property_data['ZIP']))[0]
print("Number of missing zips after second round imputation:  ", len(missing_zips))
```

```python
# Data exclusions and filling in missings:

city_owner_list = ['PARKCHESTER PRESERVAT', 
'PARKS AND RECREATION',
'DCAS',
'HOUSING PRESERVATION',
'CITY OF NEW YORK',
'DEPT OF ENVIRONMENTAL',
'BOARD OF EDUCATION',
'NEW YORK CITY HOUSING',
'CNY/NYCTA',
'NYC HOUSING PARTNERSH',
'DEPARTMENT OF BUSINES',    
'DEPT OF TRANSPORTATIO',   
'MTA/LIRR',                  
'PARCKHESTER PRESERVAT',
'MH RESIDENTIAL 1, LLC',
'LINCOLN PLAZA ASSOCIA',
'UNITED STATES OF AMER',
'U S GOVERNMENT OWNRD',
'THE CITY OF NEW YORK',
'NYS URBAN DEVELOPMENT',
'NYS DEPT OF ENVIRONME',
'CULTURAL AFFAIRS',
'DEPT OF GENERAL SERVI',
'DEPT RE-CITY OF NY']

property_data = property_data[~property_data['OWNER'].isin(city_owner_list)].copy()

print("Number of govertment owned properties that have been excluded: ", len(df) - len(property_data))
```

### Mean values of columns per tax class to use for imputation:

```python
property_data['FULLVAL'].replace(np.NaN, 0)
temp = property_data[property_data['FULLVAL'] != 0]
mean_fullval = temp.groupby('TAXCLASS')['FULLVAL'].mean()
```

```python
property_data['AVLAND'].replace(np.NaN, 0)
temp = property_data[property_data['AVLAND'] != 0]
mean_avland = temp.groupby('TAXCLASS')['AVLAND'].mean()
```

```python
property_data['AVTOT'].replace(np.NaN, 0)
temp = property_data[property_data['AVTOT'] != 0]
mean_avtot = temp.groupby('TAXCLASS')['AVTOT'].mean()
```

```python
for entry in mean_fullval.index:
    property_data.loc[(property_data['FULLVAL'] == 0) &  (property_data['TAXCLASS'] == entry), 'imputation'] = 'fullval'
    property_data.loc[(property_data['FULLVAL'] == 0) &  (property_data['TAXCLASS'] == entry), 'FULLVAL'] = mean_fullval[entry]
    
    property_data.loc[(property_data['AVLAND'] == 0) &  (property_data['TAXCLASS'] == entry), 'imputation'] = 'avland'
    property_data.loc[(property_data['AVLAND'] == 0) &  (property_data['TAXCLASS'] == entry), 'AVLAND'] = mean_avland[entry]
    
    property_data.loc[(property_data['AVTOT'] == 0) &  (property_data['TAXCLASS'] == entry), 'imputation'] = 'avtot'
    property_data.loc[(property_data['AVTOT'] == 0) &  (property_data['TAXCLASS'] == entry), 'AVTOT'] = mean_avtot[entry]
```

```python
# Stories

print("Number of records with missing #stories: ", len(property_data[property_data['STORIES'].isnull()]))
```

```python
temp = property_data[property_data['STORIES'] != 0]
mean_stories = temp.groupby('TAXCLASS')['STORIES'].mean()
```

```python
# Imputing missing stories:

property_data['STORIES'] = property_data['STORIES'].fillna(0)
for entry in mean_stories.index:
    property_data.loc[(property_data['STORIES'] == 0) &  (property_data['TAXCLASS'] == entry), 'imputation'] = 'stories'
    property_data.loc[(property_data['STORIES'] == 0) &  (property_data['TAXCLASS'] == entry), 'STORIES'] = mean_stories[entry] 
```

```python
hit_list = [0, 1, np.NaN, None]

mean_LTFRONT = property_data[~ property_data['LTFRONT'].isin(hit_list)].groupby('TAXCLASS')['LTFRONT'].mean()
mean_LTDEPTH = property_data[~ property_data['LTDEPTH'].isin(hit_list)].groupby('TAXCLASS')['LTDEPTH'].mean()
mean_BLDFRONT = property_data[~ property_data['BLDFRONT'].isin(hit_list)].groupby('TAXCLASS')['BLDFRONT'].mean()
mean_BLDDEPTH = property_data[~ property_data['BLDDEPTH'].isin(hit_list)].groupby('TAXCLASS')['BLDDEPTH'].mean()
```

```python
for entry in mean_LTFRONT.index:
    property_data.loc[(property_data['LTFRONT'].isin(hit_list)) &  (property_data['TAXCLASS'] == entry), 'imputation'] = 'ltfront'
    property_data.loc[(property_data['LTFRONT'].isin(hit_list)) &  (property_data['TAXCLASS'] == entry), 'LTFRONT'] = mean_LTFRONT[entry]
    
    property_data.loc[(property_data['LTDEPTH'].isin(hit_list)) &  (property_data['TAXCLASS'] == entry), 'imputation'] = 'ltdepth'
    property_data.loc[(property_data['LTDEPTH'].isin(hit_list)) &  (property_data['TAXCLASS'] == entry), 'LTDEPTH'] = mean_LTDEPTH[entry]

    property_data.loc[(property_data['BLDFRONT'].isin(hit_list)) &  (property_data['TAXCLASS'] == entry), 'imputation'] = 'bldfront'    
    property_data.loc[(property_data['BLDFRONT'].isin(hit_list)) &  (property_data['TAXCLASS'] == entry), 'BLDFRONT'] = mean_BLDFRONT[entry]
    
    property_data.loc[(property_data['BLDDEPTH'].isin(hit_list)) &  (property_data['TAXCLASS'] == entry), 'imputation'] = 'blddepth'
    property_data.loc[(property_data['BLDDEPTH'].isin(hit_list)) &  (property_data['TAXCLASS'] == entry), 'BLDDEPTH'] = mean_BLDDEPTH[entry]
```

### Feature Creation:

```python
property_data['ZIP'] = property_data['ZIP'].astype('int32')
```

```python
property_data['ZIP3'] = property_data['ZIP'].astype(str).str[:3]
```

```python
property_data['LTSIZE'] = property_data['LTFRONT'] * property_data['LTDEPTH']
property_data['BLDSIZE'] = property_data['BLDFRONT'] * property_data['BLDDEPTH']
property_data['BLDVOL'] = property_data['BLDSIZE'] * property_data['STORIES']
```

```python
property_data['r1'] = property_data['FULLVAL'] / property_data['LTSIZE']
property_data['r2'] = property_data['FULLVAL'] / property_data['BLDSIZE']
property_data['r3'] = property_data['FULLVAL'] / property_data['BLDVOL']
property_data['r4'] = property_data['AVLAND'] / property_data['LTSIZE']
property_data['r5'] = property_data['AVLAND'] / property_data['BLDSIZE']
property_data['r6'] = property_data['AVLAND'] / property_data['BLDVOL']
property_data['r7'] = property_data['AVTOT'] / property_data['LTSIZE']
property_data['r8'] = property_data['AVTOT'] / property_data['BLDSIZE']
property_data['r9'] = property_data['AVTOT'] / property_data['BLDVOL']
```

```python
ninevars = ['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7', 'r8', 'r9']
zip5_mean = property_data.groupby('ZIP')[ninevars].mean()
zip3_mean = property_data.groupby('ZIP3')[ninevars].mean()
taxclass_mean = property_data.groupby('TAXCLASS')[ninevars].mean()
borough_mean = property_data.groupby('B')[ninevars].mean()

property_data = property_data.join(zip5_mean, on = 'ZIP', rsuffix = '_zip5')
property_data = property_data.join(zip3_mean, on = 'ZIP3', rsuffix = '_zip3')
property_data = property_data.join(taxclass_mean, on = 'TAXCLASS', rsuffix = '_taxclass')
property_data = property_data.join(borough_mean, on = 'B', rsuffix = '_boro')

rsuffix = ['_zip5', '_zip3', '_taxclass', '_boro']

for var in ninevars:
    for r in rsuffix:
        property_data[str(var) + r] = property_data[var] / property_data[str(var) + r]
```

### Standardising data

```python
dataset = property_data.loc[:, 'r1':'r9_boro']

dataset = (dataset - dataset.mean() )/ dataset.std()

dataset
```

### Principal components analysis

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 0.99, svd_solver = 'full')
pca.fit(dataset)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.yticks(np.arange(0.05, 1.1, step = 0.1))
plt.xticks(np.arange(0, 20, step = 2))
plt.axvline(x=6, linestyle = '--')
plt.ylim(0.35,1)
```

First 6 principal components together explain 91% of the variance 

```python
pca = PCA(n_components = 6)
princ_comps = pca.fit_transform(dataset)
data_pca = pd.DataFrame(princ_comps, columns = ['PC' + str(i) for i in range(1, pca.n_components_+1)])
data_pca.shape
```

### Standardising the rotated data

```python
pca_mean = data_pca.mean()
pca_sd = data_pca.std()
data_pca = (data_pca - pca_mean) / pca_sd
data_pca.describe()
```

### Model 1: z-score outliers

```python
p = 2
one_over_p = 1/p

score1 = (data_pca.abs()**p).sum(axis=1)**one_over_p
```

```python
data_pca
```

### Model 2: autoencoder reproduction error

```python
from keras.models import Model
from keras.layers import Input, Dense
data_in = Input(shape = 6,)

encoded_dim = 3
encoded = Dense(encoded_dim, activation = 'tanh')(data_in)

decoded = Dense(6, activation = 'tanh')(data_in)

autoencoder = Model(data_in, decoded)

autoencoder.compile(optimizer = 'adam', loss = 'mean_squared_error')

autoencoder.fit(data_pca, data_pca, epochs = 3, shuffle = True, validation_data = (data_pca, data_pca))

pca_out = autoencoder.predict(data_pca)

error = pca_out - data_pca

p = 2

score2 = ((error.abs()**p).sum(axis = 1))**one_over_p

score2
```

```python
scores = pd.concat([property_data[['RECORD', 'imputation']].reset_index(drop= True), score1], axis = 1).rename(columns = {0: 'score1'})
scores['score2'] = score2
scores['score1_rank'] = scores.score1.rank()
scores['score2_rank'] = scores.score2.rank()
scores['final'] = (scores.score1_rank + scores.score2_rank) / 2
scores['review_flag'] = np.where((pca_out > data_pca).all(axis = 1), "undervalued", None)
```
